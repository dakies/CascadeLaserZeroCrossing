```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook



X, Y = [], []
for line in open('kalman_c_orig.txt', 'r'):
  values = [float(s) for s in line.split()]
  Y.append(values[0])
plt.plot(Y, label='orig')
maxy = np.max(Y)

X, Y = [], []
for line in open('kalman_filt.txt', 'r'):
  values = [float(s) for s in line.split()]
  Y.append(values[0])
plt.plot(Y, label='filt')
    
X, Y = [], []
for line in open('kalman_c_hp.txt', 'r'):
  values = [float(s) for s in line.split()]
  Y.append(values[0])
plt.plot( maxy*np.array(Y))


plt.ion()
plt.legend()
plt.show()
```
